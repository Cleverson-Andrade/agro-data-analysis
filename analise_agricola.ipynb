### üìä Agrupamento e M√©dias / Grouping and Averages
**Pergunta / Question:** Qual cultura teve a maior m√©dia de produtividade hist√≥rica? / Which crop has the highest historical average yield?

```python
# Agrupamento por cultura / Grouping by crop
produtividade_media = data.groupby("crop")["crop_yield_hg_per_ha"].mean().round(2).sort_values(ascending = False)
print(produtividade_media)
```

### üå°Ô∏è An√°lise de Extremos / Extremes Analysis
**Pergunta / Question:** Qual foi o ano com a maior temperatura m√©dia e qual a produtividade global desse ano? / What was the year with the highest average temperature and the global yield for that year?

```python
data
```

```python
# Localiza√ß√£o do valor m√°ximo de temperatura / Locating max temperature value
data.mean_air_temperature_c.max()
```

```python
# Filtragem do ano espec√≠fico / Filtering the specific year
data[data["mean_air_temperature_c"] == 30.65]
```

```python
# M√©dia global de produtividade em 2013 / Global yield average in 2013
data[data["year"] == 2013]["crop_yield_hg_per_ha"].mean().round(2)
```

### üìà An√°lise de Tend√™ncias / Trend Analysis
**Objetivo / Goal:** Visualiza√ß√£o da evolu√ß√£o da produtividade m√©dia global. / Visualizing global average yield evolution.

```python
import matplotlib.pyplot as plt
```

```python
# Agrupamento temporal / Time grouping
produtividade_por_ano = (
    data.groupby("year")["crop_yield_hg_per_ha"].mean().round(2))
```

```python
# Configura√ß√£o do gr√°fico / Plot configuration
plt.figure(figsize=(10,5))

plt.plot(
    produtividade_por_ano.index,
    produtividade_por_ano.values
)

plt.xlabel("Ano")
plt.ylabel("Produtividade m√©dia (hg/ha)")
plt.title("Evolu√ß√£o da Produtividade Agr√≠cola Global")

plt.show()
```
